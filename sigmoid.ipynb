```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Standard sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Custom sigmoids
def sigmoid1(x):
    return 1 / (1 + np.exp(-(0.5 * x + 3)))

def sigmoid2(x):
    return 1 / (1 + np.exp(-(5 * x + 11)))

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values
y_sigmoid = sigmoid(x)
y_sigmoid1 = sigmoid1(x)
y_sigmoid2 = sigmoid2(x)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_sigmoid, label='Standard Sigmoid: 1/(1+exp(-x))', linewidth=2)
plt.plot(x, y_sigmoid1, label='Sigmoid1: 1/(1+exp(-(0.5*x+3)))', linewidth=2)
plt.plot(x, y_sigmoid2, label='Sigmoid2: 1/(1+exp(-(5*x+11)))', linewidth=2)

# Add probability=0.5 line
plt.axhline(y=0.5, color='red', linestyle='--', label='Probability = 0.5')

#formatting
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Sigmoid Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-0.1, 1.1)
plt.show()
```
